### Data Cleaning 

```python
# Make a copy of the clean data
movies = df.copy()
```

#### Filling the empty string values

```python
movies.isnull().sum()
```

```python
movies.fillna("empty", inplace=True)
```

```python
movies.isnull().sum()
```

#### Removing the duplicate

```python
movies.drop_duplicates(inplace=True)
```

```python
# Just to check
movies.duplicated().sum()
```

#### Fixing the `id` column

```python
movies['id'].dtype
```

```python
movies['id'] = movies['id'].astype(str)
```

```python
# To check
movies['id'].dtype
```

#### Fixing the `release_date` column

```python
movies['release_date'].dtype
```

```python
movies['release_date'].sample()
```

The given order is month/day/year. We can take the year from the `release_year` column

```python
# Removing the last 2 characters (which indicate the year) and adding the year from the release_year column
movies['release_date'] = movies['release_date'].map(lambda x: x[:-2])
movies['release_date'] = movies['release_date'] + movies['release_year'].astype(str)
```

```python
movies['release_date'].sample()
```

```python
# Changing the type into datetime
pd.to_datetime(movies['release_date'], format='%m/%d/%Y')
```

#### Drop the unnecessary columns

```python
to_be_dropped_columns = ['imdb_id', 'budget_adj', 'revenue_adj', 'overview', 'tagline']
movies.drop(to_be_dropped_columns, axis=1, inplace=True)
# Just to check
movies.shape
```

## Exploratory Data Analysis


### Movies Analysis

#### *Addressed Question:* Which movies are the best?

```python
#Creating a bar plot
plt.rcdefaults()
fig, ax = plt.subplots()

#Data to be plotted
top_movies = movies.nlargest(5, ['popularity']).original_title
top_popularities = movies.popularity.sort_values(ascending=False).head()
y_pos = np.arange(len(top_movies))

#And now, for the graph...
ax.barh(y_pos, top_popularities, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(top_movies)
ax.invert_yaxis() 
ax.set_xlabel('Movie Popularity')
ax.set_title('Most Popular Movies')

plt.show();

fig.set_size_inches((15, 8), forward=False)
fig.savefig("Imgs/MoviesByPopularity.png")
```

```python
the_70s_movies = movies.query('release_year >= 1970 and release_year <= 1979')['original_title'].value_counts().head()
the_80s_movies = movies.query('release_year >= 1980 and release_year <= 1989')['original_title'].value_counts().head()
the_90s_movies = movies.query('release_year >= 1990 and release_year <= 1999')['original_title'].value_counts().head()
the_2000s_movies = movies.query('release_year >= 2000 and release_year <= 2017')['original_title'].value_counts().head()

fig, axs = plt.subplots(2, 2, figsize=(16,16))

axs[0, 0].bar(the_2000s_movies.keys(), the_2000s_movies)
axs[0, 0].set_title('The 2000s')

axs[0, 1].bar(the_90s_movies.keys(), the_90s_movies)
axs[0, 1].set_title('The 90s')

axs[1, 0].bar(the_80s_movies.keys(), the_80s_movies)
axs[1, 0].set_title('The 80s')

axs[1, 1].bar(the_70s_movies.keys(), the_70s_movies)
axs[1, 1].set_title('The 70s')

# Draw but don't show the plot
plt.draw()
for ax in fig.axes:
    # Get current labels
    labels = ax.get_xticklabels()
    # New labels, replacing the space with a line break
    new_labels = [x.get_text().replace(' ', '\n') for x in labels]
    # Set tick labels
    ax.set_xticklabels(new_labels)
    
# Add a main title
plt.suptitle("Most Popular Movies, By Decade",
            fontsize=16, weight='bold',
            y=1.01)

plt.tight_layout()
plt.show();
fig.savefig('Imgs/MoviesByYear.png')
```

```python
#Creating a bar plot
plt.rcdefaults()
fig, ax = plt.subplots()

#Data to be plotted
top_movies = movies.nlargest(5, ['vote_average']).original_title
top_votes = movies.vote_average.sort_values(ascending=False).head()
y_pos = np.arange(len(top_movies))

#And now, for the graph...
ax.barh(y_pos, top_votes, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(top_movies)
ax.invert_yaxis() 
ax.set_xlabel('Votes Average')
ax.set_title('Movies')

plt.show();

fig.set_size_inches((12, 8), forward=False)
fig.savefig("Imgs/MoviesByValueCount.png")
```

```python
movie_rating = []
movie_rating.append(movies.query('vote_average >= 8.5').shape[0])
movie_rating.append(movies.query('vote_average < 8.5 and vote_average >= 6').shape[0])
movie_rating.append(movies.query('vote_average < 6 and vote_average >= 4.5').shape[0])
movie_rating.append(movies.query('vote_average < 4.5').shape[0])
movie_rating
```

### *Addressed Question:* Does the movie budget affect its popularity?

```python
# Sort movies by popularity, and select the top 10
top_pop_movies = movies.sort_values(by='popularity', ascending=False).iloc[:10]
# Set the movie title as an index
top_pop_movies.set_index('original_title', inplace=True)
# Scale the budget
top_pop_movies['budget'] = top_pop_movies['budget']/10e6
```

```python
top_pop_movies[['popularity', 'budget']].plot.barh(figsize=(10,10),
                                                  width=0.5,
                                                  color=['g', 'y'])

plt.xlabel('Movies')
plt.ylabel('Popularity/Budget ($10M)')
plt.title('Budget & Popularity')

plt.legend();
fig.savefig('Imgs/BudgetVsPopularity.png')
```

#### Conclusions:
> As you can see from the graph, there's no certain correlation between movie popularity and its budget. As the popularity of the movie increases, the budget does not has a certain pattern, it sometimes increases and sometimes decreases.

### *Addressed Question:* Does the movie runtime affect its popularity?

```python
top_pop_movies[['popularity', 'runtime']].plot.bar(figsize=(10,10),
                                                  width=0.5,
                                                  color=['m', 'y'])

plt.xlabel('Movies')
plt.ylabel('Popularity/Runtime (minutes)')
plt.title('Runtime & Popularity')
plt.xticks(rotation=45)

plt.tight_layout()
plt.legend();
fig.savefig('Imgs/RuntimeVsPopularity.png')
```

#### Conclusions:
> As you can see from the graph, there's no certain correlation between movie popularity and its runtime. As the popularity of the movie increases, the runtime is almost the same. It does not has a certain pattern, it sometimes increases and sometimes decreases.

### Genres Analysis

```python
# Splitting the genres column and creating a new dataframe to use for genres analysis
genres_df = movies.assign(genres=movies['genres'].str.split('|')).explode('genres')
genres_df.head(10)
```

### *Addressed Question:* Which genres are most popular?

```python
top_genres = genres_df.genres.value_counts().sort_values(ascending=False).head(7)

plt.rcdefaults()
fig = plt.figure()
plt.bar(top_genres.keys(), top_genres)
plt.xticks(rotation=45, fontsize=8)
plt.title("Top Genres of All time")

plt.show();
fig.set_size_inches((8, 12), forward=False)
fig.savefig('Imgs/Genres.png')
```

```python
the_70s_genres = genres_df.query('release_year >= 1970 and release_year <= 1979')['genres'].value_counts().head()
the_80s_genres = genres_df.query('release_year >= 1980 and release_year <= 1989')['genres'].value_counts().head()
the_90s_genres = genres_df.query('release_year >= 1990 and release_year <= 1999')['genres'].value_counts().head()
the_2000s_genres = genres_df.query('release_year >= 2000 and release_year <= 2017')['genres'].value_counts().head()

fig, axs = plt.subplots(2, 2, figsize=(16,16))

axs[0, 0].bar(the_2000s_genres.keys(), the_2000s_genres)
axs[0, 0].set_title('The 2000s')

axs[0, 1].bar(the_90s_genres.keys(), the_90s_genres)
axs[0, 1].set_title('The 90s')

axs[1, 0].bar(the_80s_genres.keys(), the_80s_genres)
axs[1, 0].set_title('The 80s')

axs[1, 1].bar(the_70s_genres.keys(), the_70s_genres)
axs[1, 1].set_title('The 70s')

# Draw but don't show the plot
plt.draw()
for ax in fig.axes:
    # Get current labels
    labels = ax.get_xticklabels()
    # New labels, replacing the space with a line break
    new_labels = [x.get_text().replace(' ', '\n') for x in labels]
    # Set tick labels
    ax.set_xticklabels(new_labels)
    
# Add a main title
plt.suptitle("Genres of Movies, By Decade",
            fontsize=16, weight='bold',
            y=1.01)

plt.tight_layout()
plt.show();
fig.savefig('Imgs/GenresByYear.png')
```

#### Conclusions
> The top genres differ slightly over time. On average the best genres are:
    - Comedy
    - Drama
    - Thriller

### *Addressed Question:* Do certain genres get a higher budget than others?

```python
highest_budget_genres = genres_df.groupby(['genres']).mean().sort_values(ascending=False, by='budget').head()['budget']

fig = plt.figure()
plt.bar(highest_budget_genres.keys(), highest_budget_genres/10e6, align='center')
plt.xticks(fontsize=8)
plt.title("Highest Budget Genres")
plt.xlabel("Genre")
plt.ylabel("Budget ($Million)")
plt.show();

fig.set_size_inches((10, 20), forward=False)
fig.savefig('Imgs/HighBudgetGenres.png')
```

#### Conclusions
> Although Comedy, Drama, and Thrillers are the top genres, they don't have the highest budgets. On Average, Adventure, Fantasy, Action, Science Fiction, and Family movies have the highest budgets.

### Revenue & Budget Analysis

### *Addressed Question:* How does the movie revenues change every year?

```python
revenues_across_years = movies.groupby(['release_year']).sum()['revenue']

x = revenues_across_years.keys()
y = revenues_across_years
colors = np.random.rand(len(revenues_across_years))

fig = plt.figure()
plt.scatter(x, y, c=colors, alpha=0.5)
plt.show();
fig.savefig('Imgs/Revenues.png')
```

#### Conclusions
> Movie revenues increase each year as apparent in the graph. This means that the movie industry is a growing one.

### *Addressed Question:* How does the movie budget change every year?

```python
budgets_across_years = movies.groupby(['release_year']).sum()['budget']

x = budgets_across_years.keys()
y = budgets_across_years
colors = np.random.rand(len(budgets_across_years))

fig = plt.figure()
plt.scatter(x, y, c=colors, alpha=0.5)
plt.show();
fig.savefig('Imgs/Budgets.png')
```

#### Conclusions
> Movie budgets increase each year as apparent in the graph. This means that the movie industry is a growing one, and people care to spend money on it

### *Addressed Question:* What are the budgets & Revenues of the most popular genres?

```python
genres_df[genres_df.genres.isin(top_genres.index)].groupby('genres')[['revenue', 'budget']].sum().plot.barh(figsize=(10,10),
                                                                                                           width=0.5,
                                                                                                           color=['b', 'y']);
```

#### Conclusions
> The graph shows that the best and most popular genres have high budgets and revenues. As you can see, popularity doesn't have much to do with budgets, but the revenues is affected positively, which makes sense.

### Limitations of the dataset

- The dataset has movies from the year 1960 until the year 2015, which is a poor and incomplete sample of all the movies

- Only 0.1% of the dataset has average vote above 8.5

- The dataset regards the cast of the movie as one string

- The dataset contains many empty values in `director` , `production_companies`, `keywords`, and `tagline` columns, which were hard to fill and needed further data collecting.

- 75% of the budgets and revenues are zero values
